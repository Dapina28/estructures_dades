# Exercici 3
**Crea una llista amb nombres desordenats i respon a les següents preguntes:**

* Quants números hi ha?
* Quantes vegades apareix el número 3.
* Quantes vegades apareixen els nombres 3 i 4?
* Quin és el número més gran?
* Quins són els 3 números més petits?
* Quin és el rang d’aquesta llista?

```python
num = [2,5,8,1,4,3,8,9]
print(f"Hi ha {len(num)} números")
```

```python
y = 0
for x in num:
    if x == 3:
        y += 1
print(f'El número 3 apareix {y} vegades')
```

```python
y = 0
z = 0
for x in num:
    if x == 3:
        y += 1
    elif x == 4:
        z += 1
print(f'El número 3 apareix {y} vegades i el número 4, {z} vegades')
```

```python
num.sort()
print(f"El número més gran és el {num[-1]}")
```

```python
print(f"Els tres números més petits són: {num[:3]}")
```

```python
print(f"El rang de la llista és de {num[0]} a {num[-1]}")
```
